```python
import requests
import pandas as pd
import numpy as np
```

```python
#The first dataset is the raw data provided FPL and made available by the official FPL API
```

```python
#Getting FPL Data
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
#Use the requests package to make a GET request from the API endpoint:
r = requests.get(url)
```

```python
json = r.json()
#Looking at the json keys, and then we’ll create our dataframe(s)
json.keys()
```

```python
#Create Dataframes from the keys need
elements_df = pd.DataFrame(json['elements'])
elements_types_df = pd.DataFrame(json['element_types'])
teams_df = pd.DataFrame(json['teams'])
```

```python
#made a copy of the elements_df just taking the columns I thought will be usefull.
fpl_data = elements_df[['id','first_name','second_name','form','team','now_cost','minutes','value_season','total_points','event_points','bonus','goals_scored','assists','goals_conceded','own_goals','clean_sheets','saves','yellow_cards','red_cards','value_form','points_per_game','selected_by_percent','element_type','photo']]
```

```python
teams_df.head()
```

```python
elements_df.head()
```

```python
elements_types_df.head()
```

```python
#map the player position name from the elements_type_df to the fpl_data df
#Mapped the element_type column from the fpl_data data frame to singular_name of the elements_types_df
fpl_data['position'] = fpl_data.element_type.map(elements_types_df.set_index('id').singular_name)
fpl_data.head()
```

```python
#Add the team name. We can do that by mapping “name” from our teams_df:
fpl_data['team'] = fpl_data.team.map(teams_df.set_index('id').name)
#fpl_data['table_position'] = fpl_data.team.map(teams_df.set_index('id').position)
fpl_data.head()
```

```python
fpl_data.info()
```

```python
#Concatinating the first name and second name columns
fpl_data['name'] = fpl_data['first_name']+' '+fpl_data['second_name']
```

```python
#The value_season column which represents the now_cost/total_points has a data type object 
#This should be a float
#Converting it to float data type
fpl_data['value_season'] = fpl_data.value_season.astype(float)
```

```python
fpl_data.head()
```

```python
#The now_cost column doesn't display the decimal as shown on the FPL site and makes the player price look higher
#Dividing now_cost by 10 to get the actual prices of players 
#We create a column called price
fpl_data['price']= fpl_data['now_cost']/10
```

```python
fpl_data.head()
```

```python
#Creating a new column called ppmillion(i.e points per million) that contains rows from the value_season column as float datatype
#The value_season is derived by total_points/price. this Gives a view of the return on investment for each player per season
fpl_data['pp_million'] = fpl_data.value_season.astype(float)
fpl_data.head()
```

```python
#Another metric I would like to look at is the points per million per 90 mins. To adjust for the minutes played this season
fpl_data['pp_mil_per_90'] = (fpl_data['pp_million']/fpl_data['minutes'])*90
fpl_data.head()
```

```python
#Converting the point_per_game to float
fpl_data['points_per_game'] = fpl_data.points_per_game.astype(float)
```

```python
#Renaming even_points column to gw_points(i.e lates gameweek points)
fpl_data['gw_points'] = fpl_data['event_points']
```

```python
#Dropping the columns we don't need anymore
fpl_data=fpl_data.drop(['value_season','first_name','now_cost','event_points'], axis = 1)
```

```python
#The next dataset are generated by probalistic models to give expected points(xPts) and expected minutes(xMins) per gameweek
#This data is made available for free on https://thefplkiwi.github.io/webpage/
#Different xPs and xMs datasets for fpl are generated and published by different sources
```

```python
#csv url for the dataset with Expected points(xPts) and Expected minutes(xMins)
fpl_kiwi_url = 'https://raw.githubusercontent.com/theFPLkiwi/webpage/main/data/Projected_FPL_2223.csv'
```

```python
#Read data to pandas
fpl_kiwi_df=pd.read_csv(fpl_kiwi_url)
```

```python
#Preview the dataset
fpl_kiwi_df.head()
```

```python
fpl_kiwi_df.info()
```

```python
#Print out all 163 columns
for x in (fpl_kiwi_df.columns):
    print(x)
```

```python
#The columns needed are the player IDs, names, numbered columns from the latest gameweek to gw 38 representing expected mins
#Columns with GW + 0.3 represent the expected mins e.g: columns 38.3 represent the xPts for gw 38.
```

```python
#Create a list of the column names to get the latest GW
cols_lst = fpl_kiwi_df.columns
#The latest GW is the column at the 10th column index
latest_gw = int(cols_lst[9])
#Getting the xP and xMins for multiple gws allows us to plan for multiple gws ahead
#I plan based on projected xP & xMins for just 5 gws since the projected pts are measured against the actual & updated after every gw
#No of gws can be increased to the full 38 gws if desired
no_gws = latest_gw + 5 #i.e current gw + next 4gws + 1 to get the range of the next 5 gws
```

```python
#Create a dataframe with just the columns need. List comprehension is used to get the columns from the latest to final GW
fpl_kiwi_df = fpl_kiwi_df[['ID','Name']+[str(xMins) for xMins in range(latest_gw,no_gws)]+[str(xPts + 0.3) for xPts in range(latest_gw,no_gws)]]
```

```python
fpl_kiwi_df.head()
```

```python
#Renaming the columns properly to avoid confusion
#Create a dictionary of the column names and the names to replaced
#use dictionary comprehension to create a dictionary of the xMins columns
xMins= {str(n): n for n in range(latest_gw,no_gws)}
```

```python
#Update the dict values with the names to replace the columns with
for mins in xMins:
    xMins[mins] = str(xMins[mins]) +'_xMins'
```

```python
xMins
```

```python
#use dictionary comprehension to create a dictionary of the xPts columns
xPts= {str(n+0.3): n for n in range(latest_gw,no_gws)}
```

```python
#Update the dict values with the names to replace the columns with
for pts in xPts:
    xPts[pts] = str(xPts[pts]) +'_xPts'
```

```python
xPts
```

```python
#replace ID with id to be used to merge with the fpl data
id_dict = {"ID": "id"}
```

```python
#Concatenate the dictionaries
column_dict = {**id_dict,**xMins,**xPts}
```

```python
#Replace the column names with the names in the column_dict
fpl_Xpts_df = fpl_kiwi_df.rename(columns=column_dict)
```

```python
fpl_Xpts_df.head()
```

```python
#Round the xMins and xPts to 2 decimal points
fpl_Xpts_df=fpl_Xpts_df.round(2)
```

```python
fpl_Xpts_df.head()
```

```python
#Verify the data types
fpl_Xpts_df.info()
```

```python
#Get total expected Mins for the next 5 GWS
fpl_Xpts_df['total_xM']= fpl_Xpts_df[(fpl_Xpts_df.columns[2])] + fpl_Xpts_df[(fpl_Xpts_df.columns[3])] + fpl_Xpts_df[(fpl_Xpts_df.columns[4])] + fpl_Xpts_df[(fpl_Xpts_df.columns[5])] + fpl_Xpts_df[(fpl_Xpts_df.columns[6])]
fpl_Xpts_df.head()
```

```python
#Get total expected Pts for the next 5 GWS
fpl_Xpts_df['total_xP']= fpl_Xpts_df[(fpl_Xpts_df.columns[7])] + fpl_Xpts_df[(fpl_Xpts_df.columns[8])] + fpl_Xpts_df[(fpl_Xpts_df.columns[9])] + fpl_Xpts_df[(fpl_Xpts_df.columns[10])] + fpl_Xpts_df[(fpl_Xpts_df.columns[11])]
fpl_Xpts_df.head()
```
